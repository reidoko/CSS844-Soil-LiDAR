```python
import open3d as o3d
from open3d import JVisualizer
```

```python
without_chunks = o3d.io.read_point_cloud("../data/scan8 None.ply")
with_chunks = o3d.io.read_point_cloud("../data/scan9 chunks 3d.ply") 

visualizer = JVisualizer()
visualizer.add_geometry(without_chunks)
visualizer.show()
```

```python
visualizer = JVisualizer()
visualizer.add_geometry(with_chunks)
visualizer.show()
```

```python
voxel_grid = o3d.geometry.VoxelGrid.create_from_point_cloud(without_chunks, voxel_size=0.005)
o3d.visualization.draw_geometries([voxel_grid])
```

```python
import numpy as np
with_chunks_array = np.asarray(with_chunks.points)
sans_chunks_array = np.asarray(without_chunks.points)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=150)
plt.scatter(with_chunks_array[:,0], with_chunks_array[:,2], s=0.1, alpha=0.5)
plt.scatter(sans_chunks_array[:,0], sans_chunks_array[:,2], s=0.1, alpha=0.5)
plt.colorbar()
```

```python
# What does standard deviation look like?
with_chunks_array[:,2].std()
```

```python
# What does standard deviation look like?
sans_chunks_array[:,2].std()
```
